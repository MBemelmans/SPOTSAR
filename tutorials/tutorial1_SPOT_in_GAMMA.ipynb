# Tutorial 1: Sub-Pixel Offset Tracking in GAMMA

## Only tested with the Dec. 2019 release of GAMMA (I ran into problems with Nov. 2022 release)

This tutorial will go over the use of GAMMA for sub-pixel offset tracking (SPOT) using decreasing window sizes in an itterative process.

1. Folder structure
2. Input data
3. Parameter file
4. Run SLC_PO_SBAS_prep.bash
5. Profit ...


### 1. Folder structure

The required folder structure is:

[your folder]
- ./geo
- ./PO_geocoding
- ./RSLC
- ./PO_SBAS_PAIRS
- - ./CCP_CCS
- - ./DISP_TXT

./geo:
will contain the full-resolution latitude/longitude information of the referance acquisition

./PO_geocoding:
will contain the geocoding infromation at the Pixel Offset spacing of r_step and a_step

./RSLC:
will contain the .rslc and .rmli off the acquisitions

./PO_SBAS_PAIRS:
will contain subfolders per image pair and contains the output folders ./CCP_CCS and ./DISP_TXT

./CCP_CCS:
will contain the cross correlation peak .ccp and cross corelattion standard deviation .ccs binary files. 

./DISP_TXT:
will contain the PO offfset estimates as text files.

the output files have a filename structure:

[satcode 1][date1] _ [satcode 2][date2] _ rwin_awin_[dataproduct]

satcode is x for TSX/TDX, c for CSK, z of PAZ

data product is .ccp or .ccs, or *_disp.txt


### Input data

For SPOTSAR you need:

- raw data or GAMMA .slc of the acquisitions
- high-resolution and accurate DEM (is it up-to-date? are there holes?)



### Parameter file

an example parameter file is given with SPOTSAR_PROC.txt

it includes the following parameters:

- data_dir.........directory containing SAR data (.slc or otherwise)
- sat_tag..........satellite abriviation to guide GAMMA with raw data ingestion
- COMMON_ref.......reference acquisition
- dem_file.........full path to DEM file
- dem_par..........full path to DEM parameter file
- demlat...........DEM oversampling factor in latitude/northing
- demlon...........DEM oversampling factor in longitude/easting
- gc_lat...........DEM oversampling factor in latitude/northing for geocoding SPOT results (currently obsolete)
- gc_lon...........DEM oversampling factor in longitude/easting for geocoding SPOT results (currently obsolete)
- RLKS............number of range looks (currently fixed at 1)
- ALKS............number of azimuth looks (currently fixed at 1)
- cropping_flag...flag for cropping (0=no cropping | 1=cropping)
- rstart..........ranging starting line for cropping
- rstop...........range stopping line for cropping
- astart..........azimuth starting line for cropping
- astop...........azimuth stopping line for cropping
- step_win_flag...flag for seperate rwin and awin determination (currently only support 1= seperate or manual)
- rwin............(obsolete)
- azwin...........(obsolete)
- rstep...........(obsolete)
- azstep..........(obsolete)

- map_rlks........pixel offset estimate spacing in range [pixels]
- map_alks........pixel offset estimation spacing in azimuth [pixels]
- init_rngwin.....initial pixel offset window size in range (for coarse offset map)
- init_aziwin.....initial pixel offset window size in azimuth (for coarse offset map)


- rngwin_PO1......224
- aziwin_PO1......108
- rngwin_PO2......182
- aziwin_PO2......88
- rngwin_PO3......140
- aziwin_PO3......68
- rngwin_PO4......99
- aziwin_PO4......48
- rngwin_PO5......58
- aziwin_PO5......28
*currently supports 5 window sizes* 


- ccpthresh.......cross-correlation peak threshold
- bwfrac..........bandwith fraction to pass through low-pass filter
- SLC_ovr.........SLC oversampling factor (1,2,4) (use 2)
- final_cpthresh..final cross correlation peak threshold

image pairs:
- ref_ORBITxxx.....[satcode][date1] example: ref_ORBIT001   c20200927
- int_ORBITxxx.....[satcode][date2] example: int_ORBIT001   c20201119





### Run SLC_PO_SBAS_prep2.bash

if everything is set up correctly you just need to run SLC_PO_SBAS_prep.bash inside the main directory (the one containing ./RSLC and ./PO_SBAS_PAIRS)

Format:

SLC_PO_SBAS_prep.bash SPOTSAR_PROC.txt [start pair] [end pair]


start pair and end pair are used to indicate what range of the image pairs listed in the parameter file should be processed. (usefull for code that terminates or fails on a specific image pair.)

The script will ask if you would like to perform reference acquisition geocoding. this is required for the first run but subsequent runs with the same parameters can opt out to reduce processing time.

The whole script does take a while depending on the number of pairs processed. the code can currently handle up to 999 image pairs within 1 batch job. (this would take weeks though...)
